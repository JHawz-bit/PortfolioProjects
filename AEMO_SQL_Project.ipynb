Forced outages have been increasing in 2017 and are the most frequent overall.

```python
%%sql
SELECT Participant_Code, Outage_Reason, Year, COUNT(*) as Total_Number_Outage_Events, 
ROUND(AVG((ABS(JULIANDAY(end_time) - JULIANDAY(start_time)))),2) as Average_Outage_Duration_In_Days
FROM AEMR_Outage_Table
WHERE status = 'Approved'
GROUP BY Participant_Code, Outage_Reason, Year
ORDER BY Total_Number_Outage_Events DESC, Outage_Reason, Year;
```

We classify a participant based off the following criteria:
<b>
<li> High Risk - On average, the participant is unavailable for > 24 Hours (1 Day)</li>
<li> Medium Risk - On average, the participant is unavailable between 12 and 24 Hours </li>
<li> Low Risk - On average, the participant is unavailable for less than 12 Hours</li> 
</b>

```python
%%sql
WITH duration AS (
    SELECT Participant_Code, Outage_Reason, Year, COUNT(*) as Total_Number_Outage_Events, 
    ROUND(AVG((ABS(JULIANDAY(end_time) - JULIANDAY(start_time)))),2) as Average_Outage_Duration_In_Days
FROM AEMR_Outage_Table
WHERE status = 'Approved'
GROUP BY Participant_Code, Outage_Reason, Year
)
SELECT *,
CASE WHEN Average_Outage_Duration_In_Days > 1 THEN "High Risk"
    WHEN Average_Outage_Duration_In_Days >= 0.5 THEN "Medium Risk"
    WHEN Average_Outage_Duration_In_Days < 0.5 THEN "Low Risk"
    END as Risk_Classification
FROM duration
ORDER BY Total_Number_Outage_Events DESC, Outage_Reason, Year;
```

Only "Forced" Outages are unplanned and therefore have the most risk. Modifying risk classification to take that into consideration.
<b>
<li> High Risk - On average, the participant is unavailable for > 24 Hours (1 Day) OR the Total Number of Outage Events > 20 </li>
<li> Medium Risk - On average, the participant is unavailable between 12 and 24 Hours OR the Total Number of Outage Events is Between 10 and 20 </li>
<li> Low Risk - On average, the participant is unavailable for less than 12 Hours OR the Total Number of Outage Events < 10 </li> 
<li> If Outage Type is not forced, then N/A
</b>


```python
%%sql
WITH duration AS (
    SELECT Participant_Code, Outage_Reason, Year, COUNT(*) as Total_Number_Outage_Events, 
    ROUND(AVG((ABS(JULIANDAY(end_time) - JULIANDAY(start_time)))),2) as Average_Outage_Duration_In_Days
FROM AEMR_Outage_Table
WHERE status = 'Approved'
GROUP BY Participant_Code, Outage_Reason, Year
)
SELECT *,
CASE WHEN Outage_Reason <> "Forced" THEN "N/A" 
    WHEN Average_Outage_Duration_In_Days > 1  OR Total_Number_Outage_Events > 20 THEN "High Risk"
    WHEN Average_Outage_Duration_In_Days >= 0.5 OR Total_Number_Outage_Events BETWEEN 10 AND 20 THEN "Medium Risk"
    WHEN Average_Outage_Duration_In_Days < 0.5 OR Total_Number_Outage_Events < 10 THEN "Low Risk"
    END as Risk_Classification
FROM duration
ORDER BY Average_Outage_Duration_In_Days DESC;
```

```python
%%sql
WITH totalOutages AS (
    SELECT year, COUNT(*) as Total_Number_Outages FROM AEMR_Outage_Table
    WHERE status = 'Approved'
    GROUP BY year
),
totalForced AS (
    SELECT year, COUNT(*) as Total_Number_Forced_Outage_Events FROM AEMR_Outage_Table
    WHERE status = 'Approved' AND Outage_Reason = "Forced"
    GROUP BY year
)
SELECT totalOutages.year, totalOutages.Total_Number_Outages, totalForced.Total_Number_Forced_Outage_Events, 
ROUND((CAST(totalForced.Total_Number_Forced_Outage_Events as FLOAT)/CAST(totalOutages.Total_Number_Outages as FLOAT)*100),2) as Pct_Outage_Forced
FROM totalOutages
INNER JOIN totalForced
ON totalOutages.year = totalForced.year
ORDER BY totalOutages.year;
```

```python
%%sql
SELECT COUNT(*) AS Total_Number_Outages, ROUND(SUM((ABS(JULIANDAY(end_time) - JULIANDAY(start_time)))),2) AS Total_Duration_In_Days,
ROUND(SUM(Energy_Lost_MW),2) AS Total_Energy_Lost, Outage_Reason, Participant_Code, Facility_Code, Year
FROM AEMR_Outage_Table
WHERE status = 'Approved'
GROUP BY Outage_Reason, Participant_Code, Facility_Code, Year
ORDER BY Total_Energy_Lost DESC, Year
```

The participants with the highest number of outages are AURICON, GW, MELK. Additionally the outage reason is Forced/Unplanned.

```python
%%sql
SELECT ROUND(AVG((ABS(JULIANDAY(end_time) - JULIANDAY(start_time)))),2) AS Avg_Duration_In_Days,
ROUND(AVG(Energy_Lost_MW),2) AS Avg_Energy_Lost, Outage_Reason, Participant_Code, Facility_Code, Year
FROM AEMR_Outage_Table
WHERE status = 'Approved' AND Outage_Reason = 'Forced'
GROUP BY Outage_Reason, Participant_Code, Facility_Code, Year
ORDER BY Avg_Energy_Lost DESC, Year
```

```python
%%sql
WITH energyLost AS (
    SELECT ROUND(AVG(Energy_Lost_MW),2) AS Avg_Energy_Lost,
    ROUND(SUM(Energy_Lost_MW),2) AS Total_Energy_Lost, 
    Outage_Reason, Participant_Code, Facility_Code, Year
    FROM AEMR_Outage_Table
    WHERE status = 'Approved' AND Outage_Reason = 'Forced'
    GROUP BY Outage_Reason, Participant_Code, Facility_Code, Year
)
SELECT Avg_Energy_Lost, Total_Energy_Lost, ROUND((Avg_Energy_Lost / Total_Energy_Lost)*100,2) AS Pct_Energy_Lost, 
Outage_Reason, Participant_Code, Facility_Code, Year
FROM energyLost
ORDER BY Total_Energy_Lost DESC
```

```python
%%sql
WITH energy AS (
SELECT Participant_Code, Facility_Code, Description_Of_Outage, ROUND(SUM(Energy_Lost_MW),2) AS Energy_Lost, 
    (SELECT ROUND(SUM(Energy_Lost_MW),2) FROM AEMR_Outage_Table WHERE status = 'Approved' AND Outage_Reason = 'Forced') AS Total_Energy_Lost
FROM AEMR_Outage_Table
WHERE status = 'Approved' AND Outage_Reason = 'Forced' AND Participant_Code IN ('AURICON', 'GW', 'MELK')
GROUP BY Participant_Code, Description_Of_Outage
)
SELECT Participant_Code, Facility_Code, Description_Of_Outage, Energy_Lost, Total_Energy_Lost, 
ROUND((Energy_Lost/Total_Energy_Lost)*100,2) AS Pct_Energy_Lost,
RANK() OVER(PARTITION BY Participant_Code ORDER BY Energy_Lost DESC) as rank
FROM energy
ORDER BY rank, Energy_Lost DESC
LIMIT 3; 
```
